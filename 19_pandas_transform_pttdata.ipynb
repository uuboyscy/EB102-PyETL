```python
import pandas as pd
import os
```

```python
flist = os.listdir('./res_gossiping')
flist
```

```python
data = []
for textfile in flist:
    if textfile == '.ipynb_checkpoints':
        continue
    with open('./res_gossiping/%s'%(textfile), 'r', encoding='utf-8') as f:
        data_str = f.read().split('---split---')[1]
        data_row = data_str.split('\n')[1:-1]
        data.append(data_row)
#         print(data_row)
#         print(data_str)
#     print('========')
data
```

```python
columns = ['推', '噓', '分數', '作者', '標題', '時間']
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def tmpfilter(s):
    result = s.split(': ')[1]
    return result
```
