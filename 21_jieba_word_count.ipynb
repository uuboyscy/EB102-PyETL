```python
import jieba
import os

jieba.load_userdict('./mydict.txt')
```

```python
flist = os.listdir('./res_gossiping')
flist
```

```python
sb = ''

for textfile in flist:
    if textfile == '.ipynb_checkpoints':
        continue
    with open('./res_gossiping/%s'%(textfile), 'r', encoding='utf-8') as f:
        sb += f.read().split('---split---')[0]

sb
```

```python
s_cut = jieba.cut(sb)
```

```python
for w in s_cut:
    print(w)
```

```python
for w in s_cut:
    print(w)
```

```python
s_cut = jieba.cut(sb)
wc = {}

for w in s_cut:
    if len(w) < 2:
        continue
    if not w in wc:
        wc[w] = 1
    else:
        wc[w] += 1

wc
```

```python
wc_list = [(w, wc[w]) for w in wc]
wc_list
```

```python
wc_list = []
for w in wc:
    tmpdata = tuple([w, wc[w]])
    wc_list.append(tmpdata)
wc_list
```

```python
wc_list.sort(key=lambda t: t[1], reverse=True)
wc_list
```
